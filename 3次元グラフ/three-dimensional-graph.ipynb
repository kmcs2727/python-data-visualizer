```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f3(x, y) = (2x^2+y^2)exp(-(2x^2+y^2))
def f3(x0, x1):
    ans = (2 * x0 ** 2 + x1 ** 2) * np.exp(-(2 * x0 ** 2 + x1 ** 2))
    return ans
```

```python
x0_n, x1_n = 9, 9
x0 = np.linspace(-2, 2, x0_n)
x1 = np.linspace(-2, 2, x1_n)
y = np.zeros((x1_n, x0_n)) # x1_n行x0_n列
for i0 in range(x0_n):
    for i1 in range(x1_n):
        y[i1][i0] = f3(x0[i0], x1[i1])
```

```python
print(np.round(y, 2))
```

## 数値を色で描画(pcolor)

```python
# グラフ描画
plt.figure(figsize=(3.5, 3))
plt.gray() # .jet(), .pink(), .bone()なども使用可能
plt.pcolor(y)
plt.colorbar()
plt.show()
```

## 関数の表面を面で描画(surface)

```python
xx0, xx1 = np.meshgrid(x0, x1)
plt.figure(figsize=(5, 3.5))
ax = plt.subplot(projection='3d')
ax.plot_surface(
    xx0,
    xx1,
    y,
    rstride=1, # 何行おきに線を引くか 
    cstride=1, # 何列おきに線を引くか
    alpha=0.3, 
    color="blue",
    edgecolors="black"
)
ax.set_zticks((0, 0.2))
ax.view_init(75, -95)
plt.show()
```

```python
# より滑らかな面で描画
x0_n, x1_n = 100, 100
x0 = np.linspace(-2, 2, x0_n)
x1 = np.linspace(-2, 2, x1_n)
y = np.zeros((x1_n, x0_n))  # x1_n行x0_n列
for i0 in range(x0_n):
    for i1 in range(x1_n):
        y[i1][i0] = f3(x0[i0], x1[i1])

xx0, xx1 = np.meshgrid(x0, x1)
plt.figure(figsize=(5, 3.5))
ax = plt.subplot(projection='3d')
ax.plot_surface(
    xx0,
    xx1,
    y,
    rstride=8, # 何行おきに線を引くか 
    cstride=8, # 何列おきに線を引くか
    alpha=0.3, 
    color="blue",
    edgecolors="black"
)
ax.set_zticks((0, 0.2))
ax.view_init(75, -95)
plt.show()
```

## 等高線で描画(contour)

```python
x0_n, x1_n = 50, 50
x0 = np.linspace(-2, 2, x0_n)
x1 = np.linspace(-2, 2, x1_n)
xx0, xx1 = np.meshgrid(x0, x1)
y = f3(xx0, xx1) # 1行でyを求められる
plt.figure(figsize=(4, 4))
cont = plt.contour(
    xx0,
    xx1,
    y,
    levels=5, # 等高線のレベル
    colors="black"
)
cont.clabel(fmt="%.2f", fontsize=8)
plt.xlabel("$x_0$", fontsize=14)
plt.ylabel("$x_1$", fontsize=14)
plt.show()
```
