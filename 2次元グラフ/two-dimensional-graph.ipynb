## 関数を描画

```python
# f(x) = (x-2)x(x+2)
def f(x):
    return (x - 2) * x * (x + 2)
```

```python
# データ生成
x = np.linspace(-3, 3, 40) 
print(np.round(x, 2))
```

```python
# グラフ描画
plt.plot(x, f(x))
plt.show()
```

## 二つの関数を一つのグラフに描画

```python
# f2(x) = (x-1)x(x+2)
def f2(x):
    return (x - 1) * x * (x + 2)
```

```python
# データ生成
x = np.linspace(-3, 3, 100)
```

```python
# グラフ描画
plt.plot(x, f(x), "lightskyblue", label="$f$")
plt.plot(x, f2(x), "tan", label="$f_2$")
plt.legend(loc="upper left") # 左上部に配置
plt.title("comparison of $f$ and $f_2$")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.xlim(-3, 3)
plt.ylim(-15, 15)
plt.grid()
plt.show()
```

## 複数のグラフを描画

```python
# グラフ描画
plt.figure(figsize=(10, 3)) #　全体サイズを設定: 横10 縦3
plt.subplots_adjust(wspace=0.5, hspace=1.5) # グラフ間隔を設定
for i in range(6):
    plt.subplot(2, 3, i + 1) # グラフ位置を指定 2行3列に分割
    plt.title(i + 1)
    plt.plot(x, f(x), "black")
    plt.ylim(-10 - 2 * i, 10 + 2 * i)
    plt.grid()
plt.show()
```
